```python
from tkinter import *
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib
```

```python
root=Tk()
root.title("crop prediction")
root.geometry("1300x500+0+0")
root.resizable(False,False)
d = pd.read_csv('simple.csv')
d.info()
df=d
df.info()
yield_df=d
dh=d.head()
dt=d.tail()
label = np.array(d['Item'])
d= d.drop('Item', axis = 1)
d=d.drop('hg/ha_yield',axis=1)
d=d.drop('Unnamed: 0',axis=1)
d= d.drop('Area', axis = 1)
d= d.drop('Year', axis = 1)
d.info()
d_list = list(d.columns)
d = np.array(d)    
models_acc={}
train_features, test_features, train_labels, test_labels = train_test_split(d, label, test_size = 0.25, random_state =42)


yield_df_onehot = pd.get_dummies(yield_df, columns=['Area',"Item"], prefix = ['Country',"Item"])
features=yield_df_onehot.loc[:, yield_df_onehot.columns != 'hg/ha_yield']
label=yield_df['hg/ha_yield']
features.head()
features = features.drop(['Year'], axis=1)
#features = features.drop(['Item'], axis=1)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
features=scaler.fit_transform(features) 

models_acc={}
train_features1, test_features1, train_labels1, test_labels1 = train_test_split(features, label, test_size = 0.25, random_state=42)
```

```python
from sklearn.metrics import r2_score
def compare_models(model):
    model_name = model.__class__.__name__
    fit=model.fit(train_features1,train_labels1)
    y_pred=fit.predict(test_features1)
    r2=r2_score(test_labels1,y_pred)
    return([model_name,r2])
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier

models = [
     KNeighborsClassifier(n_neighbors=7),
     RandomForestRegressor(n_estimators=200, max_depth=3, random_state=0),
  DecisionTreeRegressor()]

model_train=list(map(compare_models,models)) 
print(*model_train, sep = "\n")
l=["KNeighborsClassifier","RandomForestRegressor","DecisionTreeRegressor"]   
for i in model_train:    
    models_acc[i[0]]=i[1]
print(models_acc)
```
